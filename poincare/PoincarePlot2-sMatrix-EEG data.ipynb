```python
"""
Download EEG Data file from : http://epileptologie-bonn.de/cms/front_content.php?idcat=193

and open it seperate directories

        ./data/Z
        ./data/O
        ./data/N
        ./data/F
        ./data/S

"""
%matplotlib inline
import numpy as np
import random 
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
from popfex import *
```

```python
class_text = ['Healty Open Eyes',
              'Healty Closed Eyes',
              'Patient Not from the epileptogenic zone',
              'Patient from the epileptogenic zone',
              'Patient epileptic seizure activity '
             ]
```

```python
def read_data(path):
    data = pd.read_csv(path)
    return data.values
```

```python
def read_n_data(tip,n):
    tips =['Z','O','N','F','S',]
    data = []
    for i in range(n) :
        l = random.randint(1,100)
        path = './data/%s/%s%03d.txt'%(tips[tip],tips[tip],l)        
        print path
        data.append(read_data(path).flatten())
    return np.array(data)        
```

```python
def show_data(out,ndelay=3,delay=4):
    matplotlib.rcParams['figure.figsize'] = (7.0, 7.0)
    for i in range(out.shape[0]):    
        a,m = PoincareVect(out[i],ndelay)
        amat = aMatrix(a,delay)
        plt.subplot(231+i)
        plt.imshow(amat)
```
